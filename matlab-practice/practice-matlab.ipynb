# Matplotlib example work flow


```python
# 1. Import matplotlib library

%matplotlib inline
import matplotlib.pyplot as plt


# 2. Prepare the Data

x = [1,2,3,4]
y = [11,22,33,44]


# 3. Setup plot

fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# 4. Plot the data

ax.plot(x,y)

# 5. Customize the plots
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# 6. Save and show ( You show the whole figure)
fig.savefig("../matlab-practice/sample-plot.png")
```

```python
# Make a plot from a dictionary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cashew Butter": 9
}
```

```python
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Butter Prices", xlabel="Butter Names", ylabel="Prices ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

# Two options for Subplots

```python
# subplot 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5));

# Plot to each different plots
ax1.plot(x, x/2)

ax2.scatter(np.random.random(10), np.random.random(10))

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax4.hist(np.random.randn(1000));
```

```python
# Subplots Option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axes
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```
