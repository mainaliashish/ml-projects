```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

```python
heart_disease.head()
```

```python
# Create a histogram
heart_disease['age'].plot(kind="hist", bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,40));
```

```python
# Object-oriented method (Ideal method)
# heart_disease.head()
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
# ax.set_xlim([45, 100]);
```

```python
# OO method customised
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# customize
ax.set(title="Heart Disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol");

# Legend method
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle="--");
```
